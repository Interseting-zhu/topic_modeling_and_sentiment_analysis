**根据给定时间段爬取Twitter文本，保存至json文件**

```
!snscrape --jsonl --progress --since 2021-01-22 twitter-search "#covid19 until:2021-02-01 lang:en" > text-query-tweets.json
```

```
import pandas as pd

# Reads the json generated from the CLI commands above and creates a pandas dataframe
tweets_df = pd.read_json('text-query-tweets.json', lines=True) #用pandas读取json
```
